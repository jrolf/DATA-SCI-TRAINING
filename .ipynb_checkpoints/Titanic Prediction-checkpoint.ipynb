```python
Male = [1 if val=='male' else 0 for val in TRAIN['Sex']]
TRAIN['Male'] = Male


TRAIN['Age'].fillna(TRAIN['Age'].mean(), inplace=True)
From35 = [abs(age-35) for age in TRAIN['Age']]
TRAIN['From35'] = From35 

YVar  = 'Survived'
XVars = [
    'Pclass',
    'Age',
    'From35',
    'SibSp',
    'Parch',
    'Fare',
    'Male',
]

# Replace any "NaN" values with the mean of the remaining values:
for varname in XVars:
    TRAIN[varname].fillna(TRAIN[varname].mean(), inplace=True)
    
```

```python
#TRAIN.tail()
```

```python
# Create a multi-variate linear regression on the "train" subset:

y1 = TRAIN[YVar]
X1 = TRAIN[XVars]
X1 = sm.add_constant(X1)

X1.head()
```

```python
# Fit the Model:
model = sm.OLS(y1, X1).fit()
```

```python
# Check for internal validity:
model.summary()
```

```python
Male = [1 if val=='male' else 0 for val in TEST['Sex']]
TEST['Male'] = Male

TEST['Age'].fillna(TEST['Age'].mean(), inplace=True)
From35 = [abs(age-35) for age in TEST['Age']]
TEST['From35'] = From35 

# Replace any "NaN" values with the mean of the remaining values:
for varname in XVars:
    TEST[varname].fillna(TEST[varname].mean(), inplace=True)
    
    
X2 = TEST[XVars]
X2 = sm.add_constant(X2)

# Predict output values for "test" using the model trained on "train":
preds = model.predict(X2) # preds == predictions for y in the "test" subset.
```

```python
# Convert predictions in to 0's and 1's:
preds2 = [1 if val>0.47 else 0 for val in preds]
preds2
```

```python
OUTPUT = pd.DataFrame()
OUTPUT['PassengerId'] = TEST['PassengerId']
OUTPUT['Survived'] = preds2
```
